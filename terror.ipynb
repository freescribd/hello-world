```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('globalterrorismdb_0718dist_.csv',low_memory=False)
```

```python
df.head()
```

```python
col_to_keep=[]
for i,col in enumerate(df.columns):
    if df.count()[i]==181691:
        col_to_keep.append(col)
     
```

```python
df=df[col_to_keep]
df.head()
```

```python
dfc=pd.DataFrame(df.pivot_table(index=['country_txt'],aggfunc='size').sort_values(ascending=False).head(15),columns=['No of Attacks'])
dfc.reset_index(inplace=True)
```

```python
dfc
```

```python
plt.bar(dfc.country_txt,dfc['No of Attacks'])
plt.xticks(rotation=60,ha='right')
plt.show()
```
